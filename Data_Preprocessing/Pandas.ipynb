```python
my_dict = {'name': ['Ahmed', 'Ali', 'Sara', 'Mona', None , 'Omar'],
           'age': [25, None, 22, 28 , 27 ,33],
           'city': ['Cairo', 'Alex', 'Giza', 'Cairo', 'Luxor', None]
           }
dataFrame = pd.DataFrame(my_dict)
dataFrame
```

```python
dataFrame['age']
```

```python
dataFrame[dataFrame['age'] > 25]  # filter rows where age > 25 syntax is dataFrame[condition] condition -> dataFrame['age'] > 25
dataFrame[(dataFrame['age'] > 25) & (dataFrame['city'] == 'Cairo')]  # filter rows where age > 25 and city is Cairo
dataFrame[(dataFrame['age'] > 25) | (dataFrame['city'] == 'Cairo')]  # filter rows where age > 25 or city is Cairo
```

```python
dataFrame[['name', 'city']]  # select specific columns
dataFrame.loc[0:3, ['name', 'city']]  # select specific columns and rows range
dataFrame.iloc[0:3, 0:2]  # select specific columns and rows range by index
```

```python
dataFrame['salary'] = [5000, 6000, 5500, 7000, 6500, 7200]  # add new column must match the number of rows in the data frame
dataFrame
```

```python
dataFrame['age']= dataFrame['age'] + 1  # update age column by adding 1 to each value
dataFrame
```

```python
newdf = dataFrame.drop('salary',axis=1) # drop salary column axis=1 for columns axis=0 for rows inplace=True to modify the original data frame 
# if inplace=False it returns a new data frame without modifying the original one
newdf = dataFrame.drop([0,1]) # drop rows with index 0 and 1
newdf
```

```python
newdf.reset_index  # reset index after dropping rows to have continuous index
```

```python
dataFrame.isnull()  # check for null values
dataFrame.isnull().sum()  # count of null values in each column
```

```python
dataFrame['city'].fillna(dataFrame['city'].mode()[0], inplace=True) # fill null values in city column with the most frequent value mode()[0] to get the first mode returned
dataFrame['age'].fillna(dataFrame['age'].mean(), inplace=True) # fill null values in age column with the mean value
dataFrame
```

```python
# drop null rows 
dataFrame.dropna(inplace=True)  # drop rows with null values inplace=True to modify the original data frame
dataFrame
dataFrame.reset_index(inplace=True, drop=True) 
dataFrame
```

```python
dataFrame.isnull().sum()
```

```python
# Grouping and Aggregation
x = dataFrame.groupby('city' )['age'].mean()  # group by city and calculate mean age for each city
print(x)
```

```python
def double_salary(salary):
    return salary * 2

dataFrame['doubleSalary'] = dataFrame['salary'].apply(double_salary)  # apply function to salary column
# or using lambda function
dataFrame['tripleSalary'] = dataFrame['salary'].apply(lambda x: x * 3)  # apply lambda function to salary column
dataFrame
```
