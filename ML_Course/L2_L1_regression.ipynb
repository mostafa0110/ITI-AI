```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures  
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error , r2_score
```

```python
np.random.seed(42)
X = np.random.rand(100 , 1)*10
X
```

```python
Y_actual = 2.5*X**2 + 0.5*X + np.random.rand(100,1)
Y_actual 
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X , Y_actual,test_size=0.2 , random_state= 42)
```

```python
poly_features = PolynomialFeatures(degree=20)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_poly , Y_train)
```

```python
Y_train_pred = model.predict(X_train_poly)
Y_test_pred = model.predict(X_test_poly)
```

```python
# print rmse
print("Train RMSE:", root_mean_squared_error(Y_train, Y_train_pred))
print("Test RMSE:", root_mean_squared_error(Y_test, Y_test_pred))
print("Train R2 Score:", r2_score(Y_train, Y_train_pred))
print("Test R2 Score:", r2_score(Y_test, Y_test_pred))
```
