```python

for x in range(1 ,50):
    if(x<=1):
        continue
    for y in range(2 , int(x**0.5 + 1)):
        if(x % y == 0):
            break
    else:
        print(f"{x} is prime")

```

```python
# ATM withdrawal
balance = int(input("Enter Balance"))
while balance > 0:
    amount = int(input("Enter amount to withdraw"))
    if balance - amount >= 0:
        balance = balance - amount
        print(f"Current balnce is {balance}")
    else:
        print("No enough balance")

print("balance is zero")
```

```python
# factorial 
num = int(input())
factorial = num
for x in range(1,num):
    factorial *= x

print(f"factorial of {num} is {factorial}")
```

```python
# febonacci
n = int(input("Enter n"))
a , b = 0 , 1
count = 0

while count < n:
    print(a)
    a , b = b , a + b
    count += 1
```

```python
# sum of digits 
n = int(input("Enter n"))
sum = 0
while n > 0:
    sum += int(n % 10) 
    n = n / 10

print(sum)
```

```python
# Check Palindrome
my_string = "level"
for c in range(len(my_string)//2):
    if my_string[c] != my_string[-(c+1)]:
        print("Not palindrome")
        break
else:
    print("Palindrome") 

# reverse 
reverse = my_string[::-1]

```
