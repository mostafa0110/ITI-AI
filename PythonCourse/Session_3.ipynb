```python
def divide(a,b):
    #docstring  three double quotes
    """This function divides a by b and returns the quotient and remainder."""
    quotient = a // b
    remainder = a % b
    return quotient, remainder # returning multiple values

q, r = divide(7, 3)  # recieving multiple returned values
print("Quotient:", q)
print("Remainder:", r)
print(divide.__doc__)  # accessing the docstring
# can use * to ignore some returned values 
```

```python
# Lambda function
my_list = [1, 2, 3, 4, 5]
new_list = [x**2 for x in my_list]  # will apply square to each element

square = lambda x:x**2              # anonymous function apply square to single value
print(square(2))
print(new_list)

# Example of using lambda with map apply operation to each element in the list
new_list = [map(lambda x: x**2, my_list)]  # create list object by list() or use square brackets to create list directly
print(new_list)
```

```python
# recursive function
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n * factorial(n-1)
    
print(factorial(5))
```

```python
# function to calculate average of numbers in a list

def average(numbers):
    sum = 0
    for x in numbers:
        sum += x
    return sum / len(numbers)

print(average([5,10]))
    
```

```python
# find longest string

def findString (strings):
    longest = strings[0]
    
    for str in strings:
        if len(str) > len(longest):
            longest = str

    return longest,strings.index(longest)

print(findString(["aaa", "1234" ,"1234567" , "123"]))            
```

```python
# return prime numbers in list

def findPrime (numbers):
    primes = []
    for x in numbers:
        if x <= 1:
            continue
        for y in range(2, int(x**0.5) + 1):
            if x % y == 0:
                break
        else:
            primes.append(x)

    return primes

print(findPrime([2,3,4,1,11,9,6,5,7,23]))
```

```python
# character frequency in string

def freqString (my_string):

    charFreq = {}
    for c in my_string:
        charFreq[c] = charFreq.get(c,0) + 1
    return charFreq

print(freqString("aaabbcddddd"))
```

```python
# feb with recurssion 

def feb (n):
    if n == 1 or n == 0 :
        return 1
    else:
        return feb(n-1) + feb(n-2)


        


for x in range(10):
    print(feb(x), end=" ") # use end= to set what to print at the end of the output
```

```python
def fun(*args):
    sum = 0
    for n in args:
        sum += n
    average = sum / len(args)
    #print("Sum:", sum)
    #print("Average:", average)
    #print("Number:", len(args))
    return sum, average, len(args)

a,b,c = fun(10,20,30)
print(a,b,c)
a,*_ = fun(5,10,15,20)  # get first returned value
print(a)
*_,b = fun(5,10,15,20)  # get last returned value
print(b)
a,*_,c = fun(5,10,15,20)  # ignore middle returned values
print(a,c)
# can't get the middle value only or if we know the number of returned values we can use _ for unused values
_,a, _ = fun(5,10,15,20)  # get second returned value
print(a)

a = fun(5,10,15,20)  # get all returned values as tuple
print(a)
```

```python
# classes in python
class Person:
    
    def __init__(self , name="default", age = "0"): # constructor   self refers to the current object used to access attributes and methods of the class
        self.name = name  # initialize attributes
        self.age = age    

    # we can add any attributes to the class inside any method not best practice
    # private attributes by adding __ before the attribute name
    def display_info(self):
        print("Name:", self.name)
        print("Age:", self.age)

class Student:
    pass # used to indicate an empty block 
```

```python
class Product:
    def __init__ (self,name,price,quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def restock(self,q):
        self.quantity += q
        print("Current Quantity =", self.quantity)

    
    def sell(self,quantity):
        if quantity <= self.quantity:
            self.quantity -= quantity
            print("Current Quantity =", self.quantity)


P1 = Product("Tea" , 10 , 10)

P1.restock(20)
P1.sell(20)
```

```python
class Calculator:
    def add(a,b):
        return a + b

    def subtract(a,b):
        return a - b
    
    def multiply(a,b):
        return a * b
    
    def divide(a,b):
        if a != 0:
            return a / b
        
C1 = Calculator

print(C1.add(1,2))
print(C1.subtract(1,2))
print(C1.multiply(1,2))
print(C1.divide(1,2))

```

```python
class Student:
    def __init__(self , name , id):
        self.name = name
        self.id = id
        self.courses = []

    def enroll(self , course):
        self.courses.append(course)
        
    def display_info(self):
        print(f"Name : {self.name}\nId : {self.id} ")
        print("Courses:")
        for c in self.courses:
            print("- ",c)

S1 = Student("Mostafa" , 11)
S1.enroll("ai")
S1.enroll("CPP")
S1.display_info()
```

```python
class Account:
    def __init__(self,name , account_number, address,balance=0):
        self.name = name
        self.account_number = account_number
        self.address = address
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("New balance:", self.balance)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("New balance:", self.balance)
        else:
            print("Insufficient")

    def display_info(self):
        print(f"Name :{self.name}   Account number :{self.account_number}\nAddress :{self.address}   balance :{self.balance} ")

A1 = Account( "Mostafa",111,"Cairo" ,5000)
A1.deposit(400)
A1.display_info()
```

```python
class Book:
    def __init__(self , title , author):
        self.title = title
        self.author = author
    
    def display_info(self):
        print(f"Title: {self.title}  Author: {self.author}")

class Library:
    def __init__(self):
        self.books = []

    def addBook(self, book):
        self.books.append(book)

    def list(self):
        for b in self.books:
            b.display_info()

B1 = Book("Mostafa" , "Book1")
B2 = Book("Ahmed" , "Book2")
L1 = Library()

L1.addBook(B1)
L1.addBook(B2)
L1.list()
```
